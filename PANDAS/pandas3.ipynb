# **Merging and Join**

```python
data2 = pd.DataFrame({
    'Key1':[1,2,3,4,5,6],
    "Key4":[45,65,67,78,76,56],
    "Key5":[43,43,5,65,7,34]
})
```

```python
data1 = pd.DataFrame(
    {
    "Key1" : [1,2,3,4,5,6],
    "Key2": [4,5,6,7,8,9],
    "Key3":[3,4,5,6,6,7]
    }
)
```

```python
data1
```

```python
data2
```

```python
pd.merge(data1,data2)
```

```python
pd.merge(data1,data2,on="Key1",how='left')
```
